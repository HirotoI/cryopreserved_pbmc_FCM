# Feature scaling

```python
# Columns containing the dead cell marker DAPI are to be filtered.
df_ = df
df = df.drop(['DAPI-A Dead'], axis=1)
df.columns
```

## Standardisation by the Standard Scaler

```python
df = df.reset_index()
df.head()
```

```python
df_col = df[['sample_id', 'group', 'celltype']]
df = df.drop(['sample_id', 'group', 'celltype'], axis=1)
df.head()
```

```python
from sklearn.preprocessing import StandardScaler

# The columns are to be standardized.
scaling_columns = df.columns
sc = StandardScaler().fit(df[scaling_columns])
```

```python
# Create DataFrame of standardised columns.
df_sc = pd.DataFrame(sc.transform(df[scaling_columns]), columns=scaling_columns, index=df.index)
```

```python
df_sc.head()
```

```python
df_concat = pd.merge(df_col, df_sc, left_index=True, right_index=True)
df_concat.to_csv('../data/df_sc.csv', index=False)
```
